# Learn Python Data Analytics By Example: Airline Arrival Delays

```python
import numpy as np
import pandas as pd
```

```python
date = pd.read_csv("Airline_Delay_Cause.csv")
```

```python
date
```

```python
date.head()
```

```python
date.isna().sum()
```

```python
date.dropna(inplace=True)
```

```python
date.isna().sum()
```

```python
date.info()
```

```python
date.drop(["carrier","carrier_name","airport","airport_name"],axis=1,inplace=True)
```

```python
date.info()
```

```python
date
```

```python
date
```

```python
X = date.drop(["late_aircraft_delay"],axis=1,inplace=False)
Y = date["late_aircraft_delay"]
```

```python
X
```

```python
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=44, shuffle =True)

#Splitted Data

print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , Y_train.shape)
print('y_test shape is ' , Y_test.shape)
```

```python
X_train.head()
```

```python
Y_train.head()
```

```python
X_test.head()
```

```python
Y_test.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LinearRegressionModel = LinearRegression(fit_intercept=True,copy_X=True,n_jobs=-1)
```

```python
# Assuming X_train and Y_train are already defined
LinearRegressionModel.fit(X_train, Y_train)  # Fitting the model with training data
print('Linear Regression Train Score is : ' , LinearRegressionModel.score(X_train, Y_train))
```

```python
Y_pred = LinearRegressionModel.predict(X_test)
```

```python
list(Y_pred)
```

```python
list(Y_test)
```

```python
from sklearn.metrics import mean_squared_error 
```

```python
MSEValue = mean_squared_error(Y_test, Y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Squared Error Value is : ', MSEValue)
```
